```python
import gymnasium as gym
from gymnasium.envs.toy_text.frozen_lake import generate_random_map
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from PIL import Image
```

```python
time = np.arange(0, 10, 0.1);
amplitude   = np.sin(time)

from sklearn.preprocessing import minmax_scale
y = minmax_scale(amplitude, feature_range=(-3,5))

plt.plot(amplitude)
plt.plot(y)
```

```python
map_size = 30
```

```python
env = gym.make(
        "FrozenLake-v1",
        max_episode_steps=5000,
        is_slippery=False,
        render_mode="rgb_array",
        desc=generate_random_map(
            size=map_size, p=0.9, seed=123
        ),
    )
```

```python
env.reset(seed=123)[0]
plt.imshow(env.render())
plt.axis('off')
plt.show()
```

```python
from PIL import Image
```

```python
img = np.array(Image.open('30x30_mapA.bmp'))
```

```python
# board is same shape as env.desc
board = np.empty((map_size, map_size), dtype='S1')

# initalize with all holes
board[:] = b'H'

# only True cells are frozen
board[img] = b'F'

# first cell is start
board[0][0] = b'S'

# last cell is goal
board[-1][-1] = b'G'
```

```python
board.shape
```

```python
env.env.env.env.desc = board.copy()
env.reset(seed=123)[0]
plt.imshow(env.render())
plt.axis('off')
plt.show()
```
